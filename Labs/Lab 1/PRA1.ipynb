## Jupyter Notebooks Quick Guide

- A notebook is made of **cells (blocks)** that you run **one at a time**.
- Each cell can contain **code** or **markdown** (text). 
- Run a cell by pressing **▶ (Play)** or **Shift + Enter**.
- **Outputs appear immediately below the cell** that produced them. This lets you **check results step by step**, instead of running everything at once.
- If something looks wrong, you know **which step caused it**.
- Cells must be run **from top to bottom**: later cells depend on earlier ones. If you run a cell before defining a variable, it will fail.
- Use markdown cells to **explain your reasoning** between code steps. Press **▶ (Play)** after you are done editing them, or double click if you want to re-edit.
- If you want to be extra comfortable with Jupyter Notebooks, check out the main shortcuts!

```python
## 3. Functions: Mathematical Black Boxes
''' Python functions behave like math functions.
    They are important because they let us "package" code
    so we can reuse it without rewriting it each time.
'''

def f(x): # using "def" to define a function named "f" with input "x"
    return x**2 # meaning f=x^2

# Usage: to use the function, "call" it with an input value
f(3) # x=3
```

```python
## 5. Arrays = Lists of Numbers
''' AI‑generated code will often use **arrays** (lists of numbers).
    Arrays let us do math on many numbers at once.
'''

import numpy as np 
x = np.array([1, 2, 3, 4])

#Operations apply **element‑wise**:
print(x + 2)  # adds 2 to each element

# a prebuilt dedicated package lets us do many actions easily:
print(x.shape)  # shows the shape (dimensions) of the array
```

```python
## 5. For‑Loops: Repeating a Calculation
''' A **for-loop** lets us repeat a calculation multiple times,
    updating variables as we go.
'''
s = 0
for i in range(5):
    s = s + i #Read as: > Add 0, then 1, then 2, then 3, then 4 to `s`
print(s)  
```

^ Equivalent math:
$s = \sum_{i=0}^{4} i$

2. We provide you with a sample dataset `x` and `y` representing a random curve (see below). Use your riemann_sum function to estimate the area under the curve. Compare the results with two different n (number of rectangles).

```python
# !pip install matplotlib
!pip install scipy
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)
x = np.linspace(0, 5, 100)
y = np.sin(x) + 0.2*np.random.randn(100)
y = y - y.min() + 0.1


plt.plot(x, y, label='random', color='blue', linestyle='--')

a, b = 0, 5
n = 100
x_trap = np.linspace(a, b, n + 1)
y_trap = np.interp(x_trap, x, y)

trap_area = np.trapezoid(y_trap, x_trap)
print(trap_area)

for i in range(len(x_trap) - 1):
    plt.fill(
        [x_trap[i], x_trap[i], x_trap[i+1], x_trap[i+1]],
        [0, y_trap[i], y_trap[i+1], 0],
        color="blue",
        alpha=0.3,
        edgecolor="black"
    )

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Function with Riemann Sum Approximation")
plt.grid(True)

plt.show()


stats.probplot(y, dist="norm", plot=plt)
plt.title("Q–Q Plot of y vs Normal Distribution")
plt.show()
```
